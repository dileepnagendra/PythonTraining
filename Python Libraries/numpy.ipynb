```python
marks = np.random.randint(50,100,5)
print(marks)
```

```python
res = np.where(marks>90,"A",
               np.where(marks>80,"B",
                np.where(marks>70,"C","D")))
print(res)
```

```python
def grade(n):
    if n>90:return "A"
    elif n>80: return "B"
    elif n>70: return "C"
    else: return "D"
# res = np.vectorize(grade)(marks)
showgrade = np.vectorize(grade)
res =  showgrade(marks)
print(res)
```
