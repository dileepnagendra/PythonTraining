```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
days = [15,20,24,16,14,26]
```

```python
plt.figure(figsize=(5,3))
plt.plot(months,days,"o--",color = "red",lw=0.7)
plt.grid("True")
plt.title("Number of Days Attended")
plt.xlabel("Month")
plt.ylabel("Days Attended")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar(months,days,color = ["red","green","skyblue"])
plt.grid("True")
plt.title("Number of Days Attended")
plt.xlabel("Month")
plt.ylabel("Days Attended")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(days,labels = months,autopct = "%1.1f%%")
plt.show()
```
